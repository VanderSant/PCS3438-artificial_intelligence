```python
# data initialize

data_raw = [
    [0,250,36,"A"],
    [10,150,34,"B"],
    [2,90,10,"A"],
    [6,78,8,"B"],
    [4,20,1,"A"],
    [1,170,70,"B"],
    [8,160,41,"A"],
    [10,180,38,"B"],
    [6,200,45,"X"]
    ]
column = ["a1","a2","a3","Classe"]
data = np.array(data_raw)
data_list_df = pd.DataFrame(data,columns=column)
data_list_df
```

## a)
A função a ser empregada é a distancia



```python
# b)

def get_d(a1,a2,a3,y1,y2,y3):
    delta1 = np.abs(a1-y1)
    delta2 = np.abs(a2-y2)
    delta3 = np.abs(a3-y3)
    d = np.sqrt((delta1**2)+(delta2**2)+(delta3**2))
    return d

column_b = ["index","d","classe"]
data_num = 9
pesos = []
for obj_i in range(0,len(data)):
    if(obj_i == (data_num-1)):
        continue
    a1 = data_raw[data_num-1][0]
    a2 = data_raw[data_num-1][1]
    a3 = data_raw[data_num-1][2]
    y1 = data_raw[obj_i][0]
    y2 = data_raw[obj_i][1]
    y3 = data_raw[obj_i][2]
    d = get_d(a1,a2,a3,y1,y2,y3)
    peso = [obj_i,d,data[obj_i][3]]
    pesos.append(peso)
    
pesos_list_df = pd.DataFrame(np.array(pesos),columns=column_b)
pesos_list_df['d'] = pd.to_numeric(pesos_list_df['d'])
pesos_list_df_sorted = pesos_list_df.sort_values(by=['d'], ascending=True)

class_list_df = pd.DataFrame(np.array([[0,0],[0,0]]),columns=["A","B"]).drop(1)
class_list_df_knn = pd.DataFrame(np.array([[0,0],[0,0]]),columns=["A","B"]).drop(1)

for index in range(0,len(pesos_list_df_sorted.index)):
    letter = pesos_list_df_sorted.iloc[index,2] 
    class_list_df[letter][0] += (1/(pesos_list_df_sorted.iloc[index,1]))
    class_list_df_knn = pd.concat([class_list_df_knn,class_list_df])
    
class_list_df_knn = class_list_df_knn.reset_index(drop=True).reset_index(names="K-NN")
class_list_df_knn
```
